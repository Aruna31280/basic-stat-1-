```python
# Visualize 
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


for col in df.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# dimensionality reduction with pca
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
# Select only numeric features
features = df.select_dtypes(include=np.number).columns
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])
# Fit PCA
pca = PCA()
pca.fit(X_scaled)
# Scree plot 
plt.figure(figsize=(6,4))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Explained Variance')
plt.grid(True)
plt.show()
```

```python
#conclusion and insights
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

# KMeans on original data
kmeans = KMeans(n_clusters=3, random_state=42)
labels_kmeans = kmeans.fit_predict(X_scaled)

# KMeans on PCA data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
labels_pca = kmeans_pca.fit_predict(X_pca)

# Compute metrics
silhouette_original = silhouette_score(X_scaled, labels_kmeans)
dbi_original = davies_bouldin_score(X_scaled, labels_kmeans)

silhouette_PCA = silhouette_score(X_pca, labels_pca)
dbi_PCA = davies_bouldin_score(X_pca, labels_pca)

# Create comparison table
results = pd.DataFrame({
    "Dataset": ["Original Data", "PCA Reduced Data"],
    "Silhouette Score": [silhouette_original, silhouette_PCA],
    "Davies-Bouldin Index": [dbi_original, dbi_PCA]
})

print("\nðŸ“Š Clustering Performance Comparison:\n")
print(results.to_string(index=False))
```
