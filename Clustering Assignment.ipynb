```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
df = pd.read_excel("EastWestAirlines.xlsx", sheet_name=1)  
print("Dataset shape:", df.shape)

# Keep only numeric columns
numeric_df = df.select_dtypes(include=[np.number])
if numeric_df.shape[1] == 0:
    raise ValueError("No numeric columns found in dataset. Check sheet_name!")
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
mask = ~((numeric_df < (Q1 - 1.5*IQR)) | (numeric_df > (Q3 + 1.5*IQR))).any(axis=1)
df_clean = numeric_df[mask]
```

```python
#Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clean)   
print("X_scaled shape:", X_scaled.shape)
```

```python
#visualization
import matplotlib.pyplot as plt
import numpy as np
X = np.random.rand(50, 2)
labels = np.random.randint(0, 3, 50)
def plot_clusters(X, labels, title):
    plt.figure(figsize=(6, 5))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

plot_clusters(X, labels, "Sample Clusters")
```

```python
#cluster analysis and interpreyation
it is unsupervised learning technique used on selected features,to maximize similarity within clusters and difference between clusters.
methods:k-means: partition data into k based on distance work good for spherical clusters
hierarchical: builds a tree(dendrogram) it look like nest
dbscan:density based cluster,good for irregular shaped clusters and detecting outliers
```
